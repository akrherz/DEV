```python
import psycopg2
import numpy as np
POSTGIS = psycopg2.connect(database='postgis', host='iemdb', user='nobody')
cursor = POSTGIS.cursor()

data = np.zeros( (366,5) )
cursor.execute("""SELECT extract(doy from valid) as day, min(tmpc), max(tmpc), avg(tmpc), stddev(tmpc) from raob_profile p JOIN raob_flights f
 on (f.fid = p.fid) WHERE f.station in ('KMPX','KSTC') and extract(hour from valid at time zone 'UTC') = 12
 and p.pressure = 850 GROUP by day""")
for row in cursor:
    idx = int(row[0] - 1)
    data[idx,0] = row[1]
    data[idx,4] = row[2]
    data[idx,1] = row[3] - row[4]
    data[idx,2] = row[3]
    data[idx,3] = row[3] + row[4]
```

```python
POSTGIS = psycopg2.connect(database='postgis', host='iemdb', user='nobody')
cursor = POSTGIS.cursor()
d2013 = []
days = []
cursor.execute("""SELECT extract(doy from valid) as day, tmpc from raob_profile_2013 r JOIN raob_flights f on (f.fid = r.fid)
 where f.station = 'KMPX' and r.pressure = 850 and extract(hour from valid at time zone 'UTC') = 12 ORDER by day ASC""")

for row in cursor:
    days.append( row[0] )
    d2013.append( row[1] )
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

ax.plot(np.arange(1,365), data[:-2,2], zorder=2, lw=2, c='brown', label='Avg')
ax.plot(np.arange(1,365), data[:-2,0], zorder=2, label='Min', c='b')
ax.plot(np.arange(1,365), data[:-2,4], zorder=2, label='Max', c='r')
ax.plot(days, d2013, label='2013', c='k')
ax.fill_between(np.arange(1,365), data[:-2,1], data[:-2,3], color='tan', zorder=1, label='1 SD')
ax.set_xlim(0,366)
ax.set_title("KMPX/KSTC Minneapolis Rawinsonde (1948-2013)\n 850 hPa Temperature Climatology at 12 UTC")

ax.set_xticks( (1,32,60,91,121,152,182,213,244,274,305,335,365) )
ax.set_xticklabels( ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec') )
ax.grid(True)
ax.legend(ncol=5)
ax.set_ylabel("Air Temperature $^\circ$C")

ax.annotate("27 July 2013 2.2$^\circ$C", xy=(208,2.2), xycoords='data', xytext=(-50,-80), textcoords='offset points',
                bbox=dict(boxstyle="round", fc="0.8"),
                arrowprops=dict(arrowstyle="->",
                connectionstyle="angle3,angleA=90,angleB=0"))

fig.savefig('test.ps')
import iemplot
iemplot.makefeature('test')
```
